1. MLflow Model Logging-

    This notebook logs the final selected models using MLflow. We load the saved models, configure MLflow in the project directory, log each model with its parameters and metrics, and register them for organized storage.

```python
import mlflow
import os

# Path to MLFlow folder
mlflow_dir=r'C:\Users\maa\Desktop\PYTHON\My_Projects\Real Estate Price Prediction and Classification\mlruns'
os.makedirs(mlflow_dir,exist_ok=True)

mlflow.set_tracking_uri(f'file:///{mlflow_dir}')

mlflow.set_experiment('Real_Estate_Project')
```

```python
# Importing the models that were saved with joblib
import joblib

model_r=joblib.load(r'C:\Users\maa\Desktop\PYTHON\My_Projects\Real Estate Price Prediction and Classification\data\saved_models\xgb_model_r.joblib')
model_p=joblib.load(r'C:\Users\maa\Desktop\PYTHON\My_Projects\Real Estate Price Prediction and Classification\data\saved_models\rfr_model_price.joblib')
model_c=joblib.load(r'C:\Users\maa\Desktop\PYTHON\My_Projects\Real Estate Price Prediction and Classification\data\saved_models\xgb_model_clf.joblib')
```

2. Logging each model, their hyper-parameters with their respective values and their average metric scores.

```python
# Logging Appreciation Rate model
import mlflow.xgboost
with mlflow.start_run(run_name='XGB_Regressor_AppreciationRate'):
    mlflow.log_params({'model_name':'XGB_AppreciationRate',
                      'n_estimators':300,
                      'max_depth':6,
                      'learning_rate':0.05,
                      'subsample':0.8,
                      'colsample_bytree':0.7,
                      'random_state':42})
    # Logging the model metrics
    mlflow.log_metrics({'Average_MAE': 0.007379})

    # Logging the model
    mlflow.xgboost.log_model(model_r,artifact_path='xgb_regressor_r',registered_model_name='XGB_AppreciationRate')

print('Logged and Registered: XGB Regressor (Appreciation Rate)')
```

```python
# Logging Property Price prediction model
import mlflow.sklearn
with mlflow.start_run(run_name='Random_Forest_Regressor_price'):
    mlflow.log_params({'model_name':'Random_Forest_Price',
                       'n_estimators':400,
                       'random_state':42,
                       'n_jobs':-1,
                       'max_depth':20,
                       'min_samples_split':5,
                       'min_samples_leaf':2})
    
    # Logging the model metrics
    mlflow.log_metrics({'Average_MAE':6.9243})

    # Logging the model
    mlflow.sklearn.log_model(model_p,artifact_path='random_forest_price',registered_model_name='Random_Forest_Price')

print('Logged and Registered: Random Forest Regressor (Property Price)')
```

```python
# Logging Property Price prediction model
with mlflow.start_run(run_name='XGB_Classifier_InvestmentType'):
    mlflow.log_params({'model_name':'XGB_InvestmentType',
                       'n_estimators':300,
                       'learning_rate':0.05,
                       'max_depth':6,
                       'subsample':0.8,
                       'colsample_bytree':0.8,
                       'min_child_weight':3,
                       'objective':'multi:softprob',
                       'num_class':4,
                       'random_state':42,
                       'n_jobs':-1})
    
    # Logging the model metrics
    mlflow.log_metrics({'Average_weighted_f1':0.8605})

    # Logging the model
    mlflow.xgboost.log_model(model_c,artifact_path='xgb_classifier_type',registered_model_name='XGB_InvestmentType')

print('Logged and Registered: XGB Classifier (Investment Type)')
```
